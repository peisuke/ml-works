# 7.2 階層クラスタリング
- 目的
    - Irisデータを対象に階層クラスタリングを行う
- 内容
    - 特徴量間の相関の計算
    - 相関に基づいたデータのクラスタリング
- 参考
    - 本コードは以下のサイトを参考にしました
    - https://qiita.com/maskot1977/items/4db7976a987a057ad3ba

```
import pandas as pd
from sklearn import datasets
import matplotlib.pyplot as plt
from pandas import plotting
from scipy.cluster.hierarchy import linkage, dendrogram
```

```
iris = datasets.load_iris()

X = pd.DataFrame(iris.data, columns=iris.feature_names)
y = iris.target

# 数値や特徴量の確認
display(X.head())
```

```
# データの各特徴量間の相関を可視化
# 対角要素はデータのヒストグラム（数値の分布）を示す
df = pd.DataFrame(iris.data, columns=iris.feature_names)
plotting.scatter_matrix(df, figsize=(6,6), alpha=0.8, diagonal='kde')
plt.show()
```

```
# データのクラスタリング
# 数値の近いデータ同士が繰り返し接続されていき、クラスタとなる
plt.figure(figsize=(18, 4))

result1 = linkage(iris.data, 
                  metric = 'braycurtis', 
                  method = 'average')
dendrogram(result1)
plt.title("Dedrogram")
plt.ylabel("Threshold")
plt.show()
```
